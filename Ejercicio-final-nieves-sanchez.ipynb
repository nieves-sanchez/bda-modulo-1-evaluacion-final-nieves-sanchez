```python
# Creamos la clase TiendaOnline
class TiendaOnline:
    
    
    def __init__(self):
        self.inventario = []
        self.ventas_totales = 0.0
        # BONUS - Atributo clientes, diccionario vacío
        self.clientes = {}
        
        
    def agregar_producto(self, nombre, precio, cantidad):
        # Recorremos la lista inventario
        for producto in self.inventario:
            # Si encontramos coincidencia del nombre con el valor de la clave nombre del diccionario actualizamos la cantidad y el precio si hubiera cambiado
            if nombre == producto['nombre']:
                producto['cantidad'] += cantidad
                producto['precio'] = precio
                print('Producto actualizado')
                return
            
        # Si el for termina de recorrer el inventario y no encuentra coincidencias añade un nuevo diccionario a la lista con los datos que le hemos pasado
        self.inventario.append({'nombre': nombre, 'precio': precio, 'cantidad': cantidad})
        print(f'El producto {nombre} ha sido añadido al inventario')
        
        
    def ver_inventario(self):
        # Si el inventario está vacío imprimimos un mensaje avisando
        if len(self.inventario) == 0:
            print("El inventario está vacío")
            return
        
        # Recorremos el inventario e imprimimos los datos de cada producto
        for producto in self.inventario:
            print(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
            
            
    def buscar_producto(self, nombre):
        # Recorremos el inventario e imprimimos los datos del producto si encontramos coincidencia con el nombre
        for producto in self.inventario:
            if nombre == producto['nombre']:
                print(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
                return
            
        # Si termina el for sin encontrar coincidencias imprimimos un mensaje de producto no encontrado
        print('Producto no encontrado')  
        
                
    def actualizar_stock(self, nombre, cantidad):
        # Recorremos el inventario para encontrar el producto buscando coincidencias por el nombre y actualizamos el stock si la encontramos
        for producto in self.inventario:
            if nombre == producto['nombre']:
                # Primero calculamos cómo quedaría el stock actual y solo actualizamos si el resultado final es mayor o igual a cero
                stock_actual = producto['cantidad']
                stock_final = stock_actual + cantidad
                if stock_final < 0:
                    print(f"Error: el stock no puede ser negativo. El stock actual de {producto['nombre']} es {producto['cantidad']}.")
                    return
                else:
                    producto['cantidad'] += cantidad
                    print(f'El producto {nombre} ha sido actualizado')
                    return
                
        # Si termina el for y no hemos encontrado coincidencias imprimimos un mensaje avisando de que no hemos encontrado el producto
        print(f'El producto {nombre} no está en el inventario')
        
    
    def eliminar_producto(self, nombre):
        # Recorremos el inventario para buscar el producto
        for producto in self.inventario:
            # Si encontramos coincidencias borramos el producto usando remove
            if nombre == producto['nombre']:
                self.inventario.remove(producto)
                print(f'El producto {nombre} ha sido eliminado correctamente')
                return
            
        # Si el for termina y no hemos encontrado coincidencias imprimimos un mensaje avisando
        print('El producto no está en el inventario')
        
        
    def calcular_valor_inventario(self):
        # Calculamos el valor total del inventario, imprimimos el valor y lo devolvemos
        total = 0
        
        for producto in self.inventario:
            total += producto['precio'] * producto['cantidad']
        print(f'El valor total del inventario es ${total}')
        return total
    
    
    # BONUS - Método agregar clientes
    def agregar_cliente(self, nombre, email):
        # Primero con un bucle recorremos el diccionario clientes para buscar coincidencias con el nombre y el email
        for cliente in self.clientes:
            if nombre == cliente:
                print(f"El nombre del cliente {nombre} ya existe")
                return
            
            if email == self.clientes[cliente]['email']:
                print(f"El email {email} ya está registrado")
                return
            
        # Si terminamos el for sin encontrar coincidencias añadimos al cliente
        self.clientes[nombre] = {"email": email, "compras": []}
        print(f"El cliente {nombre} ha sido añadido correctamente")
        
        
    # BONUS - Método ver clientes
    def ver_clientes(self):
        # Primero revisamos que el diccionario no esté vacío
        if len(self.clientes) == 0:
            print("No hay clientes registrados")
            return
        
        # Recorremos el diccionario e imprimimos los datos
        for cliente in self.clientes:
            print(f"Nombre: {cliente}, Email: {self.clientes[cliente]['email']}")
            
            
    # BONUS - Realizar compra
    def realizar_compra(self):
        # Diccionario vacío que será el carrito de la compra
        carrito = {}
        
        nombre_cliente = input("Nombre del cliente: ").strip().capitalize()
        
        # Comprobamos que el cliente existe, comprobando que el nombre introducido coincida con alguna clave del diccionario de clientes
        if nombre_cliente not in self.clientes:
            print("Cliente no encontrado, Resgístrese primero y vuelva a intentarlo.")
            # Devolvemos 3 valores para que posteriormente, al desempaquetar los valores devueltos en este método si el cliente no existe no nos salte error
            return None, 0, {}
        
        # Usamos un bucle While True para que el cliente pueda añadir varios artículos al carrito, el bucle solo termina cuando el usuario escribe 'salir'
        while True:
            nombre_producto = input("Nombre del producto (o 'salir' para finalizar la compra): ").strip().capitalize()
            # Si el cliente escribe 'salir' salimos del bucle y pasamos a calcular el total de la compra
            if nombre_producto.lower() == "salir":
                break
            
            # Creamos producto_seleccionado y al buscar el producto dentro del inventario si lo encuentra, crea una copia de ese diccionario en la variable
            # si no lo encuentra al salir del bucle seguirá siendo None y podremos saber si existe o no el artículo
            producto_seleccionado = None
            for producto in self.inventario:
                if nombre_producto == producto['nombre']:
                    producto_seleccionado = producto
                    break
            if producto_seleccionado is None:
                print('Producto no encontrado')
                continue
            
            # Pedimos cantidad que quiere comprar y comprobamos con 'isdigit' si solo tiene números, si no es número avisamos y volvemos a pedir número, si es
            # número convertimos a 'int'
            cantidad_a_comprar = input("Cantidad: ").strip()
            if not cantidad_a_comprar.isdigit():
                print("Por favor introduce un número")
                continue
            cantidad_a_comprar = int(cantidad_a_comprar) 
            
            # Comprobamos si hay cantidad suficiente de stock en el inventario, si no hay suficiente avisamos y volvemos a pedir y si hay suficiente restamos la 
            # cantidad del stock del inventario         
            if cantidad_a_comprar > producto_seleccionado['cantidad']:
                print(f"No hay suficiente stock. Disponible: {producto_seleccionado['cantidad']}")
                continue
            producto_seleccionado['cantidad'] -= cantidad_a_comprar
            
            # Guardamos el producto en el carrito, si el producto no está en el carrito lo añadimo y si ya está solo actualizamos la cantidad a comprar
            if nombre_producto not in carrito:                
                carrito[nombre_producto] = {'precio': producto_seleccionado['precio'], 'cantidad': cantidad_a_comprar}
            else:
                carrito[nombre_producto]['cantidad'] += cantidad_a_comprar
                
        # Cuando el usuario escribe 'salir' salimos del While y pasamos a calcular el total de la cesta        
        total_compra = 0
        
        # Recorremos todo el carrito y de cada uno de ellos multiplicamos el precio por unidad del producto por la cantidad a comprar y lo acumulamos en la variable total_compra
        for producto in carrito:
            suma = carrito[producto]['cantidad'] * carrito[producto]['precio']
            total_compra += suma
            
        # Imprimimos un mensaje con el total de la compra y devolvemos una tupla con el nombre del cliente, el total de la compra y el carrito
        print(f"El total de la cesta es ${total_compra}")
        return nombre_cliente, total_compra, carrito
    
    
    # BONUS - Procesar pago
    def procesar_pago(self, total_compra):
        
        # Imprimimos el total del pago
        print(f"Total a pagar: ${total_compra}")
        
        # EL input lo convertimos a float para asegurarnos de que introduce un número válido y lo manejamos con un try/except por si introduce algo que no es un numero.
        try:
            cantidad = float(input('Cantidad entregada por el cliente: '))
        except:
            print("Error: debes introducir un número válido.")
            return
        
        # Comprobamos que la cantidad es superior al total de la compra para procesar el pago, el total de la compra lo sacamos de la variable total_compra que nos ha devuelto
        # el método realizar_compra y así evitamos errores al introducir la cantidad a pagar porque la calcula automáticamente el programa.
        if cantidad >= total_compra:
            cambio = cantidad - total_compra
            print(f"Pago realizado correctamente. Su cambio es: ${cambio}")
        else:
            print("El importe entregado es insuficiente.")
            return
        
        
    # BONUS - resgistrar_compra
    def registrar_compra(self, nombre_cliente, carrito):
        # Revisamos que el cliente exista
        if nombre_cliente not in self.clientes:
            print("Error. Cliente no encontrado")
            return
        
        # Creamos una variable para acumular el total
        total = 0
        
        # Recorremos el carrito y para cada producto en el carrito sumamos el precio por la cantidad y lo guardamos en una variable suma para luego acumularla a la variable total
        for producto in carrito:
            suma = carrito[producto]['cantidad'] * carrito[producto]['precio']
            total += suma
        print(f"La suma total de la compra es de ${total}")
        
        # Acumulamos el total de la venta a las ventas totales de la tienda
        self.ventas_totales += total
        
        # Añadimos la venta a la lista de compras del cliente
        self.clientes[nombre_cliente]['compras'].append({'productos': carrito, 'total': total})
        
    # BONUS - ver compra cliente
    def ver_compras_cliente(self, nombre_cliente):
        # Comprobamos que el cliente existe
        if nombre_cliente not in self.clientes:
            print("Error. Cliente no encontrado")
            return
        
        #Recorremos la lista de compras del cliente y mostramos el total
        for compra in self.clientes[nombre_cliente]['compras']:
            print(f"El total de la compra es {compra['total']}")
            
            #Recorremos los productos de cada compra
            for producto in compra['productos']:
                cantidad = compra['productos'][producto]['cantidad']
                precio = compra['productos'][producto]['precio']
                print(f"Producto: {producto}, Cantidad: {cantidad}, Precio: {precio}")
```

```python
# Crear tienda
tienda = TiendaOnline()
```

```python
# Agregar productos
tienda.agregar_producto("Camisa", 20, 10)
tienda.agregar_producto("Pantalón", 30, 8)
tienda.agregar_producto("Zapatos", 50, 5)
tienda.agregar_producto("Sudadera", 40, 6)
tienda.agregar_producto("Calcetines", 5, 20)
```

```python
# Agregar clientas
tienda.agregar_cliente("Ana", "ana@email.com")
tienda.agregar_cliente("Laura", "laura@email.com")
tienda.agregar_cliente("Marta", "marta@email.com")
```

```python
# Ver inventario
tienda.ver_inventario()
```

```python
# Buscar producto
tienda.buscar_producto("Camisa")
```

```python
# Buscar producto que no existe
tienda.buscar_producto("Gorra")
```

```python
# Actualizar stock sumando
tienda.buscar_producto("Camisa")
tienda.actualizar_stock("Camisa", 5)
tienda.buscar_producto("Camisa")
```

```python
# Actualizar stock restando
tienda.buscar_producto("Camisa")
tienda.actualizar_stock("Camisa", -3)
tienda.buscar_producto("Camisa")
```

```python
# Intentar dejar stock en negativo
tienda.buscar_producto("Zapatos")
tienda.actualizar_stock("Zapatos", -10)
```

```python
# Eliminar producto existente
tienda.eliminar_producto("Sudadera")
tienda.ver_inventario()
```

```python
# Eliminar producto inexistente
tienda.eliminar_producto("Sudadera")
```

```python
# Calcular valor inventario
tienda.calcular_valor_inventario()
```

```python
# Ver clientes
tienda.ver_clientes()
```

```python
# Realizar compra, procesar pago y registrar compra
nombre_cliente, total_compra, carrito = tienda.realizar_compra()

if nombre_cliente is not None and total_compra > 0:
    tienda.procesar_pago(total_compra)
    tienda.registrar_compra(nombre_cliente, carrito)
```
